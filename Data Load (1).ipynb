```python
import yfinance as yf
```

```python
import pandas as pd
import configparser
import sqlite3
```

```python
# Load configuration file
config = configparser.ConfigParser()
config.read(r"C:\Users\Renuka\OneDrive\Desktop\config1.ini")
companies = dict(config.items('Companies'))
print(companies)
```

```python
# Define database connection and cursor
conn = sqlite3.connect('finance.db')
cursor = conn.cursor()
```

```python
# Loop through companies and download data from yahoo finance
for ticker in companies.values():
    data = yf.download(ticker)
    print(data)
```

```python
# Add company name column
data['company_name'] = ticker
```

```python
# Load data into database
data.to_sql('yahoo_finance_data13', conn, if_exists='append', index=False)
```

```python
# Close database connection
conn.close()
```
